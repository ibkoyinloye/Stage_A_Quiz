```python
#Question 9
#For the fuel type coal, what is the percentage change in the fuel cost per unit burned in 
#1998 compared to 1994?
```

```python
a = df.loc[(df['fuel_type_code_pudl'] == 'coal') & (df['report_year'] == 1998), 'fuel_cost_per_unit_burned']
b = df.loc[(df['fuel_type_code_pudl'] == 'coal') & (df['report_year'] == 1994), 'fuel_cost_per_unit_burned']

percentage_change = (a.sum() - b.sum())/ (b.sum()) * 100
print(percentage_change)
```

```python
import numpy as np
import pandas as pd
import io
import requests
url="https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"
df=pd.read_csv(io.StringIO(s.decode('utf-8')))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
grp = df.groupby('report_year').agg({'fuel_cost_per_unit_delivered':'sum'})
fig = plt.figure(figsize = (20,5))
sns.barplot(grp.index, grp['fuel_cost_per_unit_delivered'])
plt.show()

```
